```python
#importar pacotes

#tratamento de dados
import pandas as pd
import numpy as np

#Extracao de dados
import yfinance as yf
```

Extração de dados

```python
#Baixar dados históricos
bitcoin = yf.download('BTC-USD', start='2014-09-17', end='2023-11-14')
sp500 = yf.download('^GSPC', start='2014-09-17', end='2023-11-14') #sp500
interest_rate = yf.download('^IRX', start='2014-09-17', end='2023-11-14') #jurosAmericano
ouro = yf.download('GC=F', start='2014-09-17', end='2023-11-14') #ouro
```
